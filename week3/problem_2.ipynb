# Consider the 2x2 matrix $A = \begin{pmatrix} 1 & a \\ a & 1 \end{pmatrix}$ where the parameter $a \in [0,2]$

## 1. Compute the eigenvalues of $A$ as functions of parameter $a$.

$det\begin{pmatrix} 1 - \lambda & a \\ a & 1 - \lambda \end{pmatrix} = 0$

$ (1 - \lambda)^2 - a^2 = 0 $

$ \pm(1 - \lambda) = \pm a $

$1 - \lambda = a \rightarrow \lambda = 1 - a$

$1 - \lambda = -a \rightarrow \lambda = 1 + a$

$\lambda - 1 = a \rightarrow \lambda = 1 + a$

$\lambda - 1 = -a \rightarrow \lambda = 1 - a$

$\lambda = 1 - a,\, 1 + a$

```python
def eigenvalues(a):
    return 1 - a, 1 + a
```

## 3. Consider the quadratic form $f(u)=u^T Au$ where u is the two-element vector describing a point in the [x, y] plane. That is, $u=\begin{pmatrix} x & y \end{pmatrix}^T$.  Write a program which computes and makes plots of f(u) while varying the parameter a.

$u^T A = \begin{pmatrix} x & y \end{pmatrix} \begin{pmatrix} 1 & a \\ a & 1 \end{pmatrix}$

$= \begin{pmatrix} x + ya & xa + y \end{pmatrix}$

$u^TAu = \begin{pmatrix} x + ya & y + xa \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

$= x(x + ya) + y(y + xa) = x^2 + xya + y^2 + yxa = x^2 + y^2 + 2xya$

#### We know matrix is singular at a = 1
$|a| = 1 \rightarrow f(u) = (x + y)^2$

This means the matrix is **positive semidefinite** because the result can't be negative, but it can be zero in the case that $x=-y$

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np

for a in [-2, -1, -.5, 0, .5, 1, 2]:
    xRange = 1
    X = np.linspace(-xRange, xRange, 100)
    Y = np.linspace(-xRange, xRange, 100)
    X, Y = np.meshgrid(X, Y)
    Z = np.square(X) + np.square(Y) + 2 * a * np.multiply(X, Y)
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, Y, Z,
                           linewidth=0, antialiased=False, label=f"a={a}", alpha=.5)
    ax.set_title(f"a={a}")
```

## 4. What happens to the quadratic form when the matrix is singular?

As said earlier, when the matrix is singular, the quadratic form returns a nonnegative value, making the matrix positive semidefinite.

$|a| < 1 \rightarrow $ all four corners point upwards. The matrix is positive definite.

$|a| > 1 \rightarrow $ two corners point up, and two corners point down. The matrix is indefinite.

$|a| = 1 \rightarrow $ two corners point up, but the other two lie flat. The matrix is positive semidefinite when the matix is singular.
