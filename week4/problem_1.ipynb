```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read in each image as a matrix.  The Matlab function “readmatrix()” will do this.

```python
cb = pd.read_csv('checkerboard.csv', header=None)
cb = np.array(cb).astype("float")

husky = pd.read_csv('side_husky.csv', header=None)
husky = np.array(husky).astype("float")
```

## Show the image.  The Matlab function “imshow()” will do this.  My results are shown below.

```python
plt.imshow(cb, cmap='gray')
plt.show()
plt.imshow(husky, cmap='gray')
plt.show()
```

## Compute the SVD of the image matrix.  Then plot the singular values vs. index.  My results are shown below.  Note that I plot the singular values twice – once on a linear scale then again on a log scale.  I do this so you can see all the singular values.

```python
def plot_svd(img, name):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.suptitle('Horizontally stacked subplots')
    # ax1.plot(x, y)
    # ax2.plot(x, -y)
    _, svd, _ = np.linalg.svd(img)
    ax1.plot(svd)
    ax1.set_ylabel("Regular Scale")
    ax2.plot(svd)
    ax2.set_ylabel("Log Scale")
    ax2.set_yscale("log")
    fig.tight_layout()
    fig.suptitle(f"SVD by Index - {name}")
    plt.show()
    return svd

cb_svd = plot_svd(cb, "Checkerboard")
husky_svd = plot_svd(husky, "Husky")
```

## Now answer the questions:

### 1. How many non-zero singular values does the checkerboard matrix have?  (Note that many of the values are very, very small.  They are round-off error so you can consider these values to be zero.)

```python
print("Non-zero singular values on the checkerboard matrix:", len([a for a in cb_svd if a > 1e-10]))
```

### 2. What is the rank of the checkerboard matrix?

The rank is 2

### 3. Why does the checkerboard matrix have this rank?

Because there are only two linearly independent columns - one where it starts black and the other where it starts white

### 4.  How many non-zero singular values does the husky matrix have?

```python
print("Non-zero singular values on the husky matrix:", len([a for a in husky_svd if a > 1e-10]))
```

### 5. What is the rank of the husky image?
The rank is 429

### 6.  Why does the husky matrix have this rank?
Because it's a sufficiently complicated image that it's nearly statistically near 0 to have column vectors that are 100% identical
